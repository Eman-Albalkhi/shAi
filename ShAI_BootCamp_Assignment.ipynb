#  1. Basic Data Exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read dataset "Salaries"
df = pd.read_csv('Salaries.csv')
df.head()
```

```python
# Identify the number of rows and columns in the dataset
rows, columns =df.shape
print("number_rows   :", rows )
print("number_columns: ", columns)
```

```python
# determine the data types of each column
print("Data types of each column:")
print(df.dtypes)
```

```python
# check for missing values in each column.
print("Missing values in each column:")
print(df.isnull().sum())
```

# 2. Descriptive Statistics

```python
#Calculate basic statistics mean, median, mode, minimum, and maximum salary

mean_salary = df['TotalPay'].mean()
median_salary = df['TotalPay'].median()
mode_salary = df['TotalPay'].mode()[0]
min_salary = df['TotalPay'].min()
max_salary = df['TotalPay'].max()

print("Mean salary    :", mean_salary)
print("Median salary  :", median_salary)
print("Mode salary    :", mode_salary)
print("Minimum salary :", min_salary)
print("Maximum salary :", max_salary)
```

```python
# determine the range of salaries
salary_range = max_salary - min_salary
print("Salary range: " ,salary_range)
```

```python
# find the standard deviation.
std_salary = df['TotalPay'].std()
print("Standard deviation of salaries:", std_salary)
```

# 3. Data Cleaning

```python
# Simple discovery of dataset for easy cleanup
df.describe()
```

 # 6. Simple Correlation Analysis

```python
# Calculate the correlation between salary and other numerical columns
all_correlations = df[['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPayBenefits']].corrwith(df['TotalPay'])

print(all_correlations)
```

```python
# Calculate the correlation coefficient between "TotalPay" and "TotalPayBenefits"
correlation_TotalPayBenefits = df['TotalPay'].corr(df['TotalPayBenefits'])

# Create a scatter plot to visualize the relationship
plt.scatter(df['TotalPay'], df['TotalPayBenefits'])
plt.xlabel('TotalPay')
plt.ylabel('TotalPayBenefits')
plt.title(f'Correlation: {correlation_TotalPayBenefits:.2f}')
plt.show()
```

As the TotalPayBenefits value increases, the salary value also increases significantly. This relationship is strong and positive if the points scattered on the chart are clearly clustered in the upper right side of the chart.

```python
# Calculate the correlation coefficient between "TotalPay" and "OtherPay"
correlation_BasePay = df['TotalPay'].corr(df['OtherPay'])

# Create a scatter plot to visualize the relationship
plt.scatter(df['TotalPay'], df['OtherPay'])
plt.xlabel('TotalPay')
plt.ylabel('OtherPay')
plt.title(f'Correlation: {correlation_BasePay:.2f}')
plt.show()
```

The correlation is 47% and when represented as a sharp angle, this indicates a moderate positive relationship between the columns TotalPay and OtherPay. In other words, as the value of one variable increases, the value of the other increases significantly, but not dramatically.
